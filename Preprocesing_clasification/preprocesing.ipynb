```python
"Drop duplicate columns"
df= df.drop(['NOMBRES.1'], axis=1)
df.head(5)
```

```python
"""Rename Columns"""
df.columns = ['ap_paterno', 'ap_materno', 'nombres', 'edad_inic_mestruacion', 'edad_inic_act_sexual', 'menopausia_despues_51anios','exposicion_radiacion', 'asma/alergia_medicacion', 'actividad_fisica', 'sobrepeso/obesidad','comentario','comentario_adic']
df.columns
```

```python
df.head(5)
```

```python
""""Join comments columns """
df['comment']= df.comentario.str.cat(df.comentario_adic, sep =' ')
df.head()
```

```python
df ['comment'] = df['comment'].str.replace('AÑO', 'anio')
df ['comment'] = df['comment'].str.replace('AÑOS', 'anios')
df ['comment'] = df['comment'].str.replace('CENO', 'seno')

!pip install unidecode
from unidecode import unidecode

# Define function to remove accents
def remove_accents(text):
    return unidecode(text)

# Apply the function to each row of the "comment" column
df['comment'] = df['comment'].apply(remove_accents)
df.head()
```

```python
"""
Lowercase all the coments
The first step is transforming the coments into lowercase to maintain a consistent flow during the NLP tasks
and text mining. For example, 'Virus' and 'virus' will be treated as two different words in any sentence, and
hence, we need to make all the words lowercase in the coments to prevent this duplication.

"""

df['comment']= df['comment'].str.lower()

"""
Remove Punctuations

For most NLP problems, punctuation does not provide additional language information. So we generally
drop it. Similarly, punctuation symbols are not crucial for sentiment analysis. They are redundant, and the
removal of punctuation before text modeling is highly recommended.

"""
df['comment']= df['comment'].str.replace('[^a-zA-Z0-9]+', " ")

df.head()
```

```python
!pip install nltk
import nltk
from nltk.corpus import stopwords
nltk.download('vader_lexicon')
nltk.download('punkt')
## NLTK 1ibrary provides the set of stop words for spanish
nltk.download ('stopwords')
stopwords = set(stopwords . words ('spanish'))
df ['comment'] = df['comment']. apply(lambda words : ' '.join (word. lower () for word in words . split() if word not in stopwords))
print(df . head ())
```

```python
df.head()
```

```python
"""
Tokenization
Tokenization is breaking down the sentence into words an paragraphs into sentences. 
These broken pieces are called tokens (sentence tokens or word tokens), which help understand
the context and create a vocabulary.

"""
import nltk
nltk.download('wordnet')
nltk.download('punkt')

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer=nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
        return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]

df['lemmatized_tokens']= df['comment']. apply(lemmatize_text)
df.head()
```

```python
""" 
Visualizing the most frequent words in the dataset 

"""
!pip install itertools
!pip install collections
!pip install matplotlib
import itertools
import collections
import matplotlib.pyplot as plt
lemmatized_tokens= list(df['lemmatized_tokens'])
token_list= list(itertools.chain(*lemmatized_tokens))
counts_no = collections.Counter(token_list)
clean_coments= pd.DataFrame(counts_no.most_common(30), columns=['palabras','conteo'])
fig, ax =plt.subplots(figsize=(8,8))
clean_coments.sort_values(by='conteo').plot.barh(x='palabras', y='conteo', ax=ax, color='blue')
ax.set_title('Palabras con mas ocurrencia en los comentarios')
plt.show()
```

```python

# Create the word list and associated scores
word_scores = {'cancer': 5, 'tumor': 4, 'mama': 3, 'seno':2, 'biopsia': 2, 'bulto': 2, 'dolor': 2, 'antecedentes': 1, 'quimioterapia': 1 }

#Go through the word list column and assign a score to each word
scores = []
for word_list in df['lemmatized_tokens']:
    score = 0
    for word in word_list:
        if word in word_scores:
            score += word_scores[word]
    scores.append(score)

# Create the "features" column with the assigned scores
df['features'] = scores

```

```python
#Asign score at the other columns

df['feature_menopausia'] = df['menopausia_despues_51anios'].replace({'SI': 1, 'NO': 0})
df['feature_radiacion'] = df['exposicion_radiacion'].replace({'SI': 1, 'NO': 0})
df['feature_alergia'] = df['asma/alergia_medicacion'].replace({'SI': 1, 'NO': 0})
df['feature_act_fisica'] = df['actividad_fisica'].replace({'SI': 1, 'NO': 0})
df['feature_sobrepeso'] = df['sobrepeso/obesidad'].replace({'SI': 1, 'NO': 0})
```

```python
# train dataset

train_df= pd.read_csv('../data/train_data.csv')
train_df.head()
```

```python
# Drop colummns without score

columnas_eliminar = ['ap_paterno', 'ap_materno', 'nombres', 'edad_inic_mestruacion',
       'edad_inic_act_sexual', 'menopausia_despues_51anios',
       'exposicion_radiacion', 'asma/alergia_medicacion', 'actividad_fisica',
       'sobrepeso/obesidad', 'comentario', 'comentario_adic', 'comment',
       'lemmatized_tokens', 'total_score']
train_df = train_df.drop(columnas_eliminar, axis=1)
train_df
```

```python
columnas_eliminar = ['ap_paterno', 'ap_materno', 'nombres', 'edad_inic_mestruacion',
       'edad_inic_act_sexual', 'menopausia_despues_51anios',
       'exposicion_radiacion', 'asma/alergia_medicacion', 'actividad_fisica',
       'sobrepeso/obesidad', 'comentario', 'comentario_adic', 'comment',
       'lemmatized_tokens']
df_test = df.drop(columnas_eliminar, axis=1)
df_test
```

```python
!pip install -U scikit-learn scipy matplotlib

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Seleccionar las características (features) relevantes para el modelo
features = ['features',	'feature_menopausia', 'feature_radiacion',	'feature_alergia',	'feature_act_fisica',	'feature_sobrepeso']

X_train = train_df[features]
y_train = train_df['cancer_probability']
X_test = df_test[features]

# Elegir un algoritmo de clasificación, como Regresión Logística, y ajustar el modelo utilizando los datos de entrenamiento
clf = LogisticRegression()
clf.fit(X_train, y_train)

# Hacer predicciones sobre los datos de prueba
y_pred = clf.predict(X_test)

# Añadir la columna de clasificación al objeto DataFrame de prueba
submission = pd.DataFrame()
submission['Id']= test_id
submission['cancer_probability'] = y_pred 

# Guardar el conjunto de datos de prueba con la columna de clasificación agregada
submission.to_csv('submission.csv', index=False)
```
