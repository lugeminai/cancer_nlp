```python
"Drop duplicate columns"
df= df.drop(['NOMBRES.1'], axis=1)
df.head(5)
```

```python
"""Rename Columns"""
df.columns = ['ap_paterno', 'ap_materno', 'nombres', 'edad_inic_mestruacion', 'edad_inic_act_sexual', 'menopausia_despues_51anios','exposicion_radiacion', 'asma/alergia_medicacion', 'actividad_fisica', 'sobrepeso/obesidad','comentario','comentario_adic']
df.columns
```

```python
df.head(5)
```

```python
df ['comentario'] = df['comentario'].str.replace('AÑO', 'anio')
df ['comentario'] = df['comentario'].str.replace('AÑOS', 'anios')
df ['comentario_adic'] = df['comentario_adic'].str.replace('AÑO', 'anio')
df ['comentario_adic'] = df['comentario_adic'].str.replace('AÑOS', 'anios')
df.head()
```

```python
"""
Lowercase all the coments
The first step is transforming the coments into lowercase to maintain a consistent flow during the NLP tasks
and text mining. For example, 'Virus' and 'virus' will be treated as two different words in any sentence, and
hence, we need to make all the words lowercase in the coments to prevent this duplication.

"""

df['comentario']= df['comentario'].str.lower()
df['comentario_adic']= df['comentario_adic'].str.lower()

"""
Remove Punctuations

For most NLP problems, punctuation does not provide additional language information. So we generally
drop it. Similarly, punctuation symbols are not crucial for sentiment analysis. They are redundant, and the
removal of punctuation before text modeling is highly recommended.

"""
df['comentario']= df['comentario'].str.replace('[^a-zA-Z0-9]+', " ")
df['comentario_adic']= df['comentario_adic'].str.replace('[^a-zA-Z0-9]+', " ")
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
## NLTK 1ibrary provides the set of stop words for spanish
nltk.download ('stopwords')
stopwords = stopwords . words ('spanish')
df ['comentario'] = df['comentario']. apply(lambda words : ' '.join (word. lower () for word in words . split() if word not in stopwords))
print(df . head ())
```
